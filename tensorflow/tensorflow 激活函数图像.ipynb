激活函数的作用：
是为了增加神经网络模型的非线性。否则没有激活函数的每层都相当于矩阵相乘。就算你叠加了若干层之后，无非还是个矩阵相乘罢了。所以你没有非线性结构的话，根本就算不上什么神经网络。

# sigmod

```python
import numpy as np
x=np.linspace(-10,10, 1000)
y=1.0/(1+np.exp(-x))
plt.plot(x,y)
```

# RELU

   max(0,x)

```python
import numpy as np        
x = np.linspace(-10,10, 1000)
y = [ i if i>0 else 0 for i in  x ]
plt.plot(x,y)
```

# tanh函数

 >tanh读作Hyperbolic Tangent

```python
import numpy as np        
x = np.linspace(-10,10, 1000)
y = (np.exp(x)- np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.plot(x,y)
```

# Softplus 函数
Softplus(x)=log(1+ex)

```python
import numpy as np        
x = np.linspace(-10,10, 1000)
y = np.log(1+ np.exp(x))
plt.plot(x,y)
```
